```python
import cv2
import tensorflow as tf

categories = ['Dog','Cat']
```

```python
import matplotlib.pyplot as plt
def prepare(filepath):
    img_size = 50
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array,(img_size,img_size))
    new_array = new_array/255.0
    plt.imshow(new_array,cmap = 'gray')
    plt.show()
    print(new_array)
    return new_array.reshape(-1, img_size, img_size, 1)
```

```python
model = tf.keras.models.load_model("CatsVsDogs")
print(model.summary())
```

```python
prediction = model.predict([prepare('TestImages/Dog_Test.jpg')])
print(prediction)
```

```python
prediction = model.predict([prepare('TestImages/Cat_Test.jpg')])
print(prediction)
```
