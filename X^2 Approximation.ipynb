```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Activation
from tensorflow.keras.callbacks import TensorBoard 
import numpy as np
import time
import numpy as np
```

```python
import random
X = (list(np.random.randint(low = 1,high=100,size=10000)))
X = np.array(X)
y = np.array([x**2 for x in X])
X = X.reshape(-1,1000)
print(X.shape)
```

```python
model = Sequential()
model.add(Dense(64, input_shape = X.shape[1:]))
model.add(Activation('relu'))

model.add(Dense(64))
model.add(Activation('relu'))

model.add(Dense(64))
model.add(Activation('relu'))

model.add(Dense(1))

model.compile(loss='mse',
             optimizer = 'adam',
             metrics = ['accuracy'])
```

```python
model.fit(X, y, epochs = 10000, validation_split= 0.1, batch_size = 256)
```

```python
print(model.predict([4,11, 20, 8, 12]))
```
